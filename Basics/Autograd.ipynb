```python
x = torch.tensor(6.7)
y = torch.tensor(0.0)
w = torch.tensor(1.0) # wt
b = torch.tensor(0.0) # bias
```

```python
#Binary Cross Entropy Loss for scalar
def binary_entropy_loss(prediction, target):
    epsilon = 1e-8
    prediction = torch.clamp(prediction,epsilon,1-epsilon)
    return -(target*torch.log(prediction)+(1-target)*torch.log(1-prediction))
```

```python
# forward pass
z = w*x + b
y_pred = torch.sigmoid(z)
loss = binary_entropy_loss(y_pred,y)
```

```python
x = torch.tensor(6.7)
y = torch.tensor(0.0)
w = torch.tensor(1.0 , requires_grad=True) # wt
b = torch.tensor(0.0 , requires_grad=True) # bias
```

```python
z = w*x + b 
y_pred =  torch.sigmoid(z)
y_pred
```

```python
loss = binary_entropy_loss(y_pred,y)
loss
```

```python
loss.backward()
```

```python
print(w.grad)
print(b.grad)
```
